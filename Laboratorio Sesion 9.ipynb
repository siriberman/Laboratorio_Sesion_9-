```python
# Parte 1
# Ejercicios de Páctica 1
#1.Se tiene un conjunto de datos sobre empleados que incluye su nombre, edad, salario y fecha de ingreso.
#Se requiere implementar algunos pasos de aseguramiento de la calidad de los datos,
#como la verificación de valores faltantes y la validación de formatos de datos

import pandas as pd

#Datos originales
data = {
'Nombre': ['Juan', 'Ana', 'Pedro', 'María', 'Luis', 'Ana'],
'Edad':[28, 4, None, 45, 38, 34],
'Salario':['3000', '4000', 'cinco mil', '4500', '4000', '5000'],
'Fecha Ingreso':['2021-01-15',	'2020/03/12',	'2022-07-01', '2021/12/01', '2021-05-20', '2020-03-12']
}
df = pd.DataFrame(data)

# 1.Identificar y contar los valores faltantes en cada columna
print("Valores faltantes:")
print(df.isnull().sum())

# 2.Verificar el formato de la columna 'Salario' para asegurar que todos los valores sean numéricos
print("\nFormato incorrecto en 'Salario'")
print(df[~df['Salario'].str.replace('{^0-9]', '').str.isdigit()])

# 3.Validar y estandarizar el formato de la columna 'Fecha Ingreso' a YYYY-MM-DD
df['Fecha Ingreso'] = pd.to_datetime(df['Fecha Ingreso'], errors='coerce', format='%Y-%m-%d')
print("\nDatos después de la validación y estandarización de 'FechaIgreso':")
print(df)
```

```python
#2.	Se tiene un conjunto de datos sobre empleados que incluye su nombre, edad, salario,
#puesto de trabajo y fecha de ingreso.
#Se requiere implementar aseguramiento de la calidad de los datos, verificando la completitud de los datos,
#la validez de los salarios y la consistencia de las fechas de ingreso.

import pandas as pd

#Datos originales
data = {
'Nombre': ['Juan', 'Ana', 'Pedro', 'María', 'Luis', 'Carla'],
'Edad':[28, None, 40, 45, 38, 34],
'Salario':['3000', '4000', '4500', 'cuatro mil', '5000', '4000'],
'Cargo':['Analista', 'Gerente', 'Desarrollador', 'Gerente', 'Analista', None],
'Fecha Ingreso': ['2021-01-15', '2020/03/12', '2022-07-01', '2021/12/01', None, '2020-03-12']
}
df = pd.DataFrame(data)

#Mostrar el DataFrame original
print("DataFrame original:")
print(df)

#Identificar y contar los valores faltantes en cada columna.
valores_faltantes = df.isnull().sum()
print("Valores faltantes por columna:")
print(valores_faltantes)

#Verificar el formato de la columna "Salario" para asegurarse de que todos los valores sean numéricos.
salarios_incorrectos = df[~df['Salario'].str.replace('{^0-9]', '').str.isdigit()]
print("\nSalarios incorrectos:")
print(salarios_incorrectos)

#Validar y estandarizar el formato de la columna "FechaIngreso" a YYYY-MM-DD..
df['Fecha Ingreso'] = pd.to_datetime(df['Fecha Ingreso'], errors='coerce', format='%Y-%m-%d')

#Verificar que no haya valores faltantes en la columna "Cargo"
Cargo_faltantes = df[df['Cargo'].isnull()]
print("\nCargo faltante:")
print(Cargo_faltantes)

#Mostrar el DataFrame actualizado
print("\nDataFrame actualizado:")
print(df)
```

```python
# Parte 2
# Ejercicios de Páctica 2
#1.	Se requiere validar los datos: Precios de Productos, se tiene una lista de productos con los respectivos precios
#y se necesita validar que ningún precio sea negativo o nulo.

#Se requiere:
#1.	Crear un DataFrame con productos y precios.
#2.	Producto: Producto A, Producto B,Producto C, Producto D.
#3. Precio:150, -25, 100, None.
import pandas as pd
data = {
    'Producto': ['Copiadora', 'Monitor', 'Computador', 'Impresora'],
    'Precio': [150, -25, 100, None]
}
df_productos = pd.DataFrame(data)
print("DataFrame original:")
print(df_productos)

#4.	Validar que todos los precios sean mayores o iguales a cero y no nulos.
df_productos['Validación Precio'] = df_productos['Precio'].apply(lambda x: 'Válido' if pd.notnull(x) and x >= 0 else 'Inválido')
print("DataFrame con validación de precios:")

#5.	Mostrar los productos con precios inválidos.
df_inválidos = df_productos[df_productos['Validación Precio'] == 'Inválido']

display(df_inválidos)
```

```python
# Parte 2
# Ejercicios de Páctica 2
#2.	Se requiere validar los datos duplicados: Clientes en una Base de Datos,
#se tiene una lista de clientes donde algunos registros tienen el mismo ID y necesitas identificarlos.

#1.	Crear un DataFrame con información de clientes.
#2. ID Cliente:1, 2, 3, 2, 4.
#3.	Nombre: Ana, Luis, María, Luis, Carlos
import pandas as pd

# Crear el DataFrame
data = {
    'ID Cliente': [1, 2, 3, 2, 4],
    'Nombre': ['Ana', 'Luis', 'María', 'Luis', 'Carlos']
}
df_clientes = pd.DataFrame(data)

#4.	Verificar si hay duplicados en la columna de ID del cliente.
duplicados = df_clientes[df_clientes.duplicated(subset=['ID Cliente'], keep=False)]

#5.	Mostrar los registros duplicados.
print("Registros duplicados:")
duplicados

```

```python
# Parte 2
# Ejercicios de Páctica 2
#3. Se requiere tener seguridad en los datos: Encriptación y Desencriptación de Datos Sensibles,
#se tiene un número de tarjeta de crédito que deseas encriptar para protegerlo y luego desencriptarlo para el uso.
#1.	Crear una cadena con datos sensibles (un número de tarjeta 1234-5678-9101- 1121).
#2.	Encriptar esos datos utilizando una clave generada.
#3.	Desencriptar los datos y mostrar ambos (encriptado y desencriptado).

from cryptography.fernet import Fernet

# 1. Crear una cadena con datos sensibles
datos_sensibles = "1234-5678-9101-1121"

# 2. Generar una clave
key = Fernet.generate_key()
f = Fernet(key)

# 3. Encriptar los datos
datos_encriptados = f.encrypt(datos_sensibles.encode())

# 4. Desencriptar los datos
datos_desencriptados = f.decrypt(datos_encriptados).decode()

# 5. Mostrar ambos
print("Datos originales:", datos_sensibles)
print("Datos encriptados:", datos_encriptados)
print("Datos desencriptados:", datos_desencriptados)

```
